```python
import matplotlib.pyplot as plt
```

```python
def judge(i):
    return i % 2 == 0

def density_a(N):
    for i in range(0, N):
        for j in range(2 ** (i), 2 ** (i + 1)):
            if judge(i):
                yield j % 2 == 0
            
            else:
                yield j % 2
```

```python
def density_b(N):
    for i in range(0, N):
        for j in range(2 ** (i), 2 ** (i + 1)):
            if judge(i):
                yield j % 2 == 0
            
            else:
                yield j % 2 and 0
```

```python
m = 10
count = 0
n = 0
data_1 = []
data_2 = []
for i in density_a(m):
    count += i
    n += 1
    data_1.append(count / n)

n = 0
count = 0

for i in density_b(m):
    count += i
    n += 1
    data_2.append(count / n)
```

```python
plt.title("N = 2 ^" + str(m) )
plt.plot(data_1)
plt.savefig("williams_exercise_016_1_"+str(m)+".png")
plt.show()
```

```python
plt.title("N = 2 ^" + str(m) )
plt.plot(data_2)
plt.savefig("williams_exercise_016_2_"+str(m)+".png")
plt.show()
```

```python
class Color:
    BLACK     = '\033[30m'
    RED       = '\033[31m'
    GREEN     = '\033[32m'
    YELLOW    = '\033[33m'
    BLUE      = '\033[34m'
    PURPLE    = '\033[35m'
    CYAN      = '\033[36m'
    WHITE     = '\033[37m'
    END       = '\033[0m'
    BOLD      = '\038[1m'
    UNDERLINE = '\033[4m'
    INVISIBLE = '\033[08m'
    REVERCE   = '\033[07m'

def twopow(count):
    while count > 1:
        if count % 2 == 0:
            count //= 2
            continue
        else:
            return False
    
    return True
```

```python
count = 0
d = 0
clr = [Color.BLUE, Color.YELLOW]
f = 1
for i in density_a(m):
    count += 1
    d += i
    if twopow(count):
        f = (f + 1) % 2
    print(clr[f], count , Color.END, 'O' if i else 'X', d/count)
```
