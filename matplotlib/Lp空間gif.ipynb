#### $L^p$空間定義
$$\|x\|_p = \left(\sum|x_i|^p\right)^{1/p}$$
#### 仕様
$x = (x_1, x_2)\in L^p\ and\ \|x\|_p = N\ \ (N \in (0, 1])$
$N$を$n\in \mathbb{N}$分割し、$\|x\|_p = N$のグラフを生成<p>

```python
"ライブラリの宣言"

#animationが動くための宣言
%matplotlib notebook

# その他宣言
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
```

```python
n = 5
m = 15
dx = 100
s, t = .5, 1
fig = plt.figure()
ims = []
def aLst(lst, n):
    return lst if n == 1 else [i * n for i in lst]

def abs(x):
    return x if x >= 0 else -x

def drawLp(p, n, dx = 100):
    figure = plt.plot()

    for j in range(n + 1):
        nlm = s + (t - s) * j/n
        yN_p = []
        yN_n = []
        x_p = [i/dx * nlm for i in range(dx + 1)]
        x_n = [-i/dx * nlm for i in range(dx + 1)]

        for i in x_p:
            res = (nlm ** p - abs(i) ** p) ** (1./p)
            yN_p.append(res)
            yN_n.append(-res)
        
        plt.gca().set_aspect('equal', adjustable='box')    
        figure += plt.plot(x_p + x_p[::-1] + x_n + x_n[::-1], yN_p + yN_n[::-1] + yN_n + yN_p[::-1])
        

    return figure

for p in range(1, m + 1):
    ims.append(drawLp(p, n, dx))

ani = animation.ArtistAnimation(fig, ims, interval = 100)
```
