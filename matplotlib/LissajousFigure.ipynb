```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
```

```python
fig = plt.figure()
ims = []
a = 2
b = 3
c = .25 * np.pi
r = [0, 2 * np.pi]
N = 100

# draw Lissajous figure
coor = [[], []]

for i in range(N + 1):
    T = i * (r[1] - r[0]) / N
    coor[0].append(np.cos(a * T))
    coor[1].append(np.sin(b * T + c))
    plt.gca().set_aspect('equal', adjustable='box')
    ims.append(plt.plot(coor[0], coor[1],  c = 'blue'))

for i in range(N):
    ims.append(plt.plot(coor[0][i:], coor[1][i:],  c = 'blue'))

ani = animation.ArtistAnimation(fig, ims, interval = 20)
```
